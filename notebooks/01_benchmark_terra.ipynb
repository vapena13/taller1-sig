```r
#Ejercicios con TERRA
# Cargar librería terra
library(terra)

# Iniciar el conteo de tiempo
t0 <- Sys.time()

# Abrir la ruta del archivo .zip en formato Virtual GDAL
s2_path <- readLines("../data/s2_shared_path.txt")

# Cargar el raster de Sentinel-2
r <- rast(s2_path)

# Seleccionar la banda 4 (roja) de la imagen
b4 <- r[[4]]

# Realizar una operación simple sobre la banda (escala)
res_terra <- b4 * 1.5

# Medir el tiempo de cálculo de la media global y asegurar que m sea numérico
m <- global(res_terra, "mean", na.rm = TRUE)[1,1]

# Calcular el tiempo total de lectura y operación
t_terra <- as.numeric(Sys.time() - t0)

# Mostrar resultados
cat("Resultados Terra:",
    round(t_terra, 3), "seg |",
    "Promedio =", round(m, 6), "\n")
```

```r
# ------------------------------------------------
# 6. Plot (FUERA DEL BENCHMARK, proxy)
# ------------------------------------------------
plot(res_terra, col = terrain.colors(100),
     main = "Terra: Banda 4 × 1.5")
```

```r
# ------------------------------------------------
# 7. Limpieza
# ------------------------------------------------
rm(r, b4, res_terra)
gc()
```
